## Problem 1

```python
import numpy as np
import matplotlib.pyplot as plt

sample_sizes = [3, 10, 50, 100, 500, 1000]
mean_sds = []

# Calculate SEM for each sample size
for N in sample_sizes:
    # Generate 1000 samples of size N from a normal distribution
    samples = np.random.normal(loc=0, scale=1, size=(1000, N))

    # Calculate mean of each sample
    means = np.mean(samples, axis=1)

    # Calculate standard deviation of sample means (SEM)
    sem = np.std(means) / np.sqrt(N)

    # Append SEM to list
    mean_sds.append(sem)

plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, mean_sds, 'o-', label='SEM')
plt.plot(sample_sizes, 1/np.sqrt(sample_sizes), 'b--', label='Theoretical SEM (1/√N)')
plt.xlabel('Sample Size')
plt.ylabel('SEM')
plt.title('SEM vs Sample Size')
plt.legend()
plt.grid(True)
plt.show()
```

# Arbitrary distribution

When $N$, the sample size, becomes large, the sample mean (denoted as $\bar{X}_N$) is distributed according to the Central Limit Theorem (CLT). The Central Limit Theorem states that regardless of the distribution of the population from which the samples are drawn, the distribution of the sample mean approaches a normal distribution as $N$ increases.

Specifically, if $X_1, X_2, \ldots, X_N$ are i.i.d. random variables with mean $\mu$ and finite variance $\sigma^2$, then for large $N$, the sample mean $\bar{X}_N = \frac{1}{N} \sum_{i=1}^{N} X_i$ is approximately normally distributed:

\[ \bar{X}_N \sim N\left(\mu, \frac{\sigma^2}{N}\right) \]

Here’s what each component means:

- $\mu$: The mean of the population from which the samples are drawn.
- $\sigma^2$: The variance of each individual random variable $X_i$.
- $\frac{\sigma^2}{N}$: The variance of the sample mean $\bar{X}_N$.

### Key Points:

1. **Normal Distribution**: The CLT guarantees that for large $N$, $\bar{X}_N$ follows a normal distribution regardless of the shape of the original population distribution (as long as the population has a finite variance).

2. **Mean and Variance**: The mean of $\bar{X}_N$ is $\mu$, which means the sample mean is an unbiased estimator of the population mean $\mu$.

3. **Standard Error**: The standard deviation of $\bar{X}_N$, also known as the standard error of the mean (SEM), decreases as $\frac{\sigma}{\sqrt{N}}$. This reflects that larger sample sizes result in more precise estimates of the population mean.

4. **Approximation**: The approximation to a normal distribution becomes more accurate as $N$ increases. For practical purposes, $N$ as low as 30 or 40 often already provides a good approximation to normality for many distributions.

In summary, when $N$ becomes large, the sample mean $\bar{X}_N$ is normally distributed around the population mean $\mu$, with a standard deviation (standard error) that decreases as $\frac{1}{\sqrt{N}}$. This fundamental property of the sample mean underpins much of inferential statistics and is crucial for understanding how sample statistics behave as sample size increases.
